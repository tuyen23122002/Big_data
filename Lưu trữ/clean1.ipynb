```python
import pandas as pd
```

```python
df = pd.read_csv('/content/estate.csv')
df.info()
```

```python
df = df[df['price'].notnull()]
df = df[df['square'].notnull()]
df = df[df['price'].astype(str).str[0].str.isdigit()]
df = df[df['square'].astype(str).str[0].str.isdigit()]
df = df[df['province'] == 'Hà Nội']
```

```python
print(len(df))
list_data = df.to_dict('records')
```

```python
for i in range(len(list_data)):
    list_data[i]['square'] = float(list_data[i]['square'].split()[0])
    price = float(list_data[i]['price'].split()[0])
    if list_data[i]['price'].split()[1] == 'tỷ':
        price *= 1000000000
    elif list_data[i]['price'].split()[1] == 'triệu':
        price *= 1000000
    list_data[i]['price'] = price
    list_data[i]['price/square'] = round(list_data[i]['price'] / list_data[i]['square'], 2)
```

```python
df_final_data = pd.DataFrame(list_data)
df_final_data.to_csv('final_data.csv')
```
